# list coprehension

```python
df["yas"]=["genc" if itr<df.age.mean() else "yasli" for itr in df.age] #:))
print(df.age.mean())
df.loc[:10,["age","yas"]]
```

```python
df.columns=[itr.upper() for itr in df.columns]
df.head()# :))
```

```python
df.columns=[itr.split("_")[0]+"#"+itr.split("_")[1] if len(itr.split("_"))>1 else itr for itr in df.columns]
df.head()
```
